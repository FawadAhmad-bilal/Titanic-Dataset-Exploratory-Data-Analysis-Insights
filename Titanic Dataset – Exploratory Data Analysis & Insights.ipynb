```python
fig, axes = plt.subplots(2, 3, figsize=(15, 9))

# Chart 1
axes[0,0].bar(["Died","Survived"],
              df["Survived"].value_counts().sort_index(),
              color=["#f87171","#4ade80"])
axes[0,0].set_title("Survival Count")
axes[0,0].grid(axis="y", alpha=0.3)

# Chart 2
gender = df.groupby("Sex")["Survived"].mean()*100
axes[0,1].bar(["Male(0)","Female(1)"],
              gender.values,
              color=["#38bdf8","#f472b6"])
axes[0,1].set_title("Survival by Gender (%)")
axes[0,1].set_ylim(0,100)
axes[0,1].grid(axis="y", alpha=0.3)

# Chart 3
pclass = df.groupby("Pclass")["Survived"].mean()*100
axes[0,2].bar(["1st","2nd","3rd"],
              pclass.values,
              color=["#facc15","#38bdf8","#f87171"])
axes[0,2].set_title("Survival by Class (%)")
axes[0,2].set_ylim(0,100)
axes[0,2].grid(axis="y", alpha=0.3)

# Chart 4
axes[1,0].hist(df[df["Survived"]==0]["Age"],
               bins=25, alpha=0.6,
               color="#f87171", label="Died")
axes[1,0].hist(df[df["Survived"]==1]["Age"],
               bins=25, alpha=0.6,
               color="#4ade80", label="Survived")
axes[1,0].set_title("Age Distribution")
axes[1,0].legend()
axes[1,0].grid(True, alpha=0.2)

# Chart 5
axes[1,1].scatter(df[df["Survived"]==0]["Age"],
                  df[df["Survived"]==0]["Fare"],
                  color="#f87171", alpha=0.4,
                  s=20, label="Died")
axes[1,1].scatter(df[df["Survived"]==1]["Age"],
                  df[df["Survived"]==1]["Fare"],
                  color="#4ade80", alpha=0.4,
                  s=20, label="Survived")
axes[1,1].set_title("Age vs Fare")
axes[1,1].legend()
axes[1,1].grid(True, alpha=0.2)

# Chart 6
embarked = df.groupby("Embarked")["Survived"].mean()*100
axes[1,2].bar(["Southampton(0)","Cherbourg(1)","Queenstown(2)"],
              embarked.values,
              color=["#f87171","#4ade80","#facc15"])
axes[1,2].set_title("Survival by Embarked (%)")
axes[1,2].set_ylim(0,100)
axes[1,2].grid(axis="y", alpha=0.3)

plt.suptitle("Titanic EDA Dashboard", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```
